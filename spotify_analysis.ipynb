```python
import pandas as pd
```

```python
df = pd.read_csv("spotify_songs.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['playlist_genre'].value_counts()
```

```python
import matplotlib.pyplot as plt
genre_counts = df['playlist_genre'].value_counts()
genre_counts.plot(kind='bar')
plt.title("Number of Songs by Genre")
plt.xlabel("Genre")
plt.ylabel("Count")
plt.show()
```

```python

avg_popularity = df.groupby('playlist_genre')['track_popularity'].mean()
print(avg_popularity)
```

```python

avg_popularity.plot(kind='bar')
plt.title("Average Popularity by Genre")
plt.xlabel("Genre")
plt.ylabel("Average Popularity")
plt.show()
```

```python

top_artists = df['track_artist'].value_counts().head(10)

print(top_artists)
```

```python

top_artists.plot(kind='bar')
plt.title("Top 10 Artists by Number of Songs")
plt.xlabel("Artist")
plt.ylabel("Number of Songs")
plt.show()
```

```python
numeric_columns = [
    'track_popularity',
    'danceability',
    'energy',
    'loudness',
    'speechiness',
    'acousticness',
    'instrumentalness',
    'valence',
    'tempo'
]

correlation_matrix = df[numeric_columns].corr()
correlation_matrix
```

```python
import seaborn as sns

plt.figure(figsize=(10,8))
sns.heatmap(correlation_matrix, annot=True)
plt.title("Correlation Heatmap of Spotify Features")
plt.show()
```
